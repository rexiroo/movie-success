## Simplify Dataset
Create new dataframe with only information needed for project while also reorganizing data for better readability.

```python
simplified_df = movie_df.filter(['id','imdb_id','original_title','title','genres','budget','revenue','release_date','runtime','vote_average','vote_count'], axis=1)
```

```python
simplified_df
```

## Cleaning Data
All rows should have usable budget data, genres, revenue data, release date, etc.

```python
simplified_df = simplified_df.dropna()
```

```python
simplified_df = simplified_df.astype({'budget':'float'}, {'runtime':'float'})
```

```python
simplified_df = simplified_df[(simplified_df != 0).all(1)]
```

```python
simplified_df
```

```python
simplified_df.info()
```

## Separate Genres into Individual Columns

```python
simplified_df['genres'] = simplified_df['genres'].str.findall('([A-Z]\w+)').transform(','.join)
```

```python
simplified_df
```

```python
simplified_df[['genre_term_1', 'genre_term_2', 'genre_term_3', 'genre_term_4', 'genre_term_5', 'genre_term_6', 'genre_term_7', 'genre_term_8', 'genre_term_9']] = simplified_df['genres'].str.split(',', expand=True)
```

```python
simplified_df
```

## Adjust for Inflation

```python
inflation_df = pd.read_csv('Datasets/inflation_data.csv')
```

```python
inflation_df.amount = round(15.49/inflation_df.amount,4)
inflation_dict = dict(zip(list(inflation_df.year), list(inflation_df.amount)))
```

```python
def adjust_for_inflation(x, column):
    date_str = str(x.release_date)
    year_str = date_str.split('-')[0]
    return round(inflation_dict[int(year_str)] * x[column],2)
```

```python
for yr in ('18','190','191','192'):
    simplified_df.drop(simplified_df[simplified_df['release_date'].str.startswith(yr)].index, inplace=True)
```

```python
simplified_df['budget_adjusted_to_2020'] = simplified_df.apply(lambda x: adjust_for_inflation(x, 'budget'),axis=1)
simplified_df['revenue_adjusted_to_2020'] = simplified_df.apply(lambda x: adjust_for_inflation(x, 'revenue'),axis=1)
```
